array reshaping

```python
#reshape array
import numpy as np
arr = np.arange(12)
up_arr = arr.reshape(3,4)
print (up_arr)

#example 2d
import numpy as np
arr1 = np.arange(14)
arr_1= np.reshape(arr1,(2,7))
print(arr_1)

#example 3d
import numpy as np

arr12 = np.arange(24)
arr_up= np.reshape(arr12,(3,2,4))
print(arr_up)
```

```python
#flatten -> makes copy of original array on updating the updated array does not change original array
import numpy as np
arr= np.arange(12).reshape(3,4)
print(arr)
arr_flat = arr.flatten()
arr_flat[0]=100 #-> does not update original array
print(arr_flat)
print(arr)

#example 3d
import numpy as np
arr1 = np.arange(24).reshape(3,2,4)
print(arr1)
arr_1= arr1.flatten()
print(arr_1)
```

```python
#ravel -> on changing updated array original array also changes->shares same memory
import numpy as np
#example 2d
arr=np.arange(12).reshape(3,4)
print(arr)
arr_up = arr.ravel()
arr_up[0]=100 #-> updates original array also
print(arr_up)
print(arr)

#example 3d
import numpy as np
arr1 = np.arange(24).reshape(3,2,4)
print(arr1)
arr_1= arr1.ravel()
print(arr_1)
```

```python
#transpose
import numpy as np
arr = np.arange(12).reshape(3,4)
print(arr)
arr_up = arr.transpose()
print(arr_up)

#example transpose 3d
import numpy as np
arr1 = np.arange(24).reshape(3,2,4)
print(arr1)
arr_1= arr1.transpose()
print(arr_1)
```

looping in numpy

```python
#while loop
import numpy as np
arr= np.arange(12)
i=0
while i < len(arr):
 # print(arr[i],end=" ")
  i += 1
# while loop 2d

arr1 = np.arange(12).reshape(3,4)

i = 0
while i < len(arr1):
    j = 0
    while j < len(arr1[i]):
        print(arr1[i][j], end=" ")
        j += 1
    print()
    print("\n")
    i += 1

#while loop for 3d
```

```python
#for loop 1d
import numpy as np
arr = np.arange(12)
for i in arr:
    print(i, end = " ")

#2d array
print("2d array")
import numpy as np
arr1 = np.arange(12).reshape(3,4)
for i in arr1:
    for j in i:
        print(j,end = " ")

#print 3d array
import numpy as np
arr2 = np.arange(24).reshape(2,3,4)
for i in arr2:
  for j in i:
    for k in j:
      print(k, end=" ")
    print()
  print()
```
