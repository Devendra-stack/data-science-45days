```python
#comparison
import numpy as np
import time


# list
start = time.time()
l = [i*2 for i in range(1000000)]
print("list output:",time.time() - start)

# array
start = time.time()
arr = np.array(1000000)*2
print("array output:",time.time() - start)
```
