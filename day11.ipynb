Data Preprocessing

```python
#handling missing value
import pandas as pd

data = {
       'Name':["dheeraj","jay","raj"],
       'Age':[25,None,32],
       'Salary':[ 5000,6000,None]
}
df = pd.DataFrame(data)
print("Original Data")
print(df)
#find null
df_null= df.isnull().sum()#gives couunt of null value per  column
print(df_null)

#dropna
df_drop=df.dropna()
print(df_drop)


#fillna
df_fill=df['Age'].mean()
print(df_fill)
df['Age'].fillna(df_fill,inplace=True)
print(df)

df_fill1 =df['Salary'].mean()
print(df_fill1)
df.fillna({'Salary': 5500}, inplace=True)
print(df)


```

encoding catogorical values

```python
#encoding
import pandas as pd
from sklearn.preprocessing import LabelEncoder
data = {
       'Name':["dheeraj","jay","raj","Neetu","Kavi","Sapu"],
       'Age':[25,None,32,20,21,22],
       'Salary':[ 5000,6000,None,7000,8000,9000],
       'Gender':['Male','Male','Male',None,'Female','Female']
}
df = pd.DataFrame(data)
print("Original Data")
print(df)
le = LabelEncoder()
df_label = df.copy()
df_label['Gender'] = df['Gender'].fillna('unknown')
df_label['Gender_encoder'] = le.fit_transform(df_label['Gender'])

print("After label encoding:")
print(df_label)
```

```python
#one hot encoding
import pandas as pd
data = {
    'colors':['red','blue','green','red','Blue']
}


df = pd.DataFrame(data)
print("original data")
print(df)
df['colors'] = df['colors'].str.lower()
print(df)

# one hot encoding
encoded_df = pd.get_dummies(df,columns=['colors'])

print("after one hot encoding")
print(encoded_df)
```
