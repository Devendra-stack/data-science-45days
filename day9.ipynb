```python
#dic
import pandas as pd
dict =  {"name":["dheeraj", "kunal","dev","raj","abhishek"],
        "age":[10,12,23,45,34],
         "salary":[{"in hand":"15000", "ctc":"1.2lpa"},{"in hand":"20000", "ctc":"2.2lpa"},
                   {"in hand":"25000", "ctc":"2.4lpa"},{"in hand":"30000", "ctc":"3.0lpa"},
                   {"in hand":"35000", "ctc":"3.2lpa"}]}
df= pd.Series(dict)

print(df)
df = pd.DataFrame(dict)
print(df)
```
